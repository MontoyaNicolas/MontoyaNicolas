```python
# Extraer las coordenadas de latitud y longitud
coordinates = gpd.GeoSeries(balt.geometry).centroid
latitudes = coordinates.y
longitudes = coordinates.x
```

```python
fig, ax = plt.subplots(figsize=(7, 7))
ax.scatter(latitudes,longitudes,c = "red")
plt.show()
```

#Mapas de Calor

```python
# Crear el gráfico
fig, ax = plt.subplots(figsize=(10, 8))

# Visualizar los datos con un gráfico de densidad 2D
balt.plot(ax=ax, color='grey', alpha=0.9, markersize=7)

# Configurar el mapa de calor (heatmap) con el gradiente de color
balt.plot(ax=ax, column='PRICE', cmap = 'coolwarm', alpha=0.4, legend=True)

# Mostrar el gráfico
plt.show()
```
